```python
# for

# for in <collection>
#       <loop body>



for v1 in range(10): # 0 ~ 9
    print('v1 is :', v1)
    
print()

for v2 in range(1, 11): # 1 ~ 10
    print('v2 is :', v2)

print()

for v3 in range(1, 11, 2):
    print('v3 is :', v3)
    
    
print()

sum1 = 0
for v in range(1, 1001):
    sum1 += v
    
print('1 ~ 1000 Sum :', sum1)


# sum() 이용
print('1 ~ 1000 Sum :', sum(range(1, 1001)))
# range() --> class<range>    내부적으로 list를 생성한다
print('1 ~ 1000 4의 배수의 합 :', sum(range(4, 1001, 4)))
```

```python
# Iteratble 자료형
# 문자열, 리스트, 튜플, 집합, 사전(딕셔너리)
# iterable 리턴 함수: range, reversed, enumerate, filter, map, zip

# 1
names = ['Kim', 'Park', 'Cho', 'Lee', 'Choi', 'Yoo']

for n in names:
    print('You are :', n)
    
# 2
lotto_numbers = [11, 19, 21, 28, 36, 37]

for n in lotto_numbers:
    print('Current numebr : ', n)
    
    
# 3
word = 'Beautiful'

for s in word:
    print('word: ', s)
    
    
# 4
my_info = {
    "name": "Lee",
    "Age": 33,
    "City": "Seoul"
}

for k in my_info:
    print('key: ',k)
    print('val: ', my_info[k])
    print('get(key): ', my_info.get(k))
    
for v in my_info.values():
    print('values :', v)
    
# 5
name = 'PineAppLE'

for n in name:
    if n.isupper():
        print(n)
    else:
        print(n.upper())
```
