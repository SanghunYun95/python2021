```python
# While


# while  <expr>:
#       <statement(s)>


# 1
n = 5
while n > 0:
    print(n)
    n = n -1

    
# 2
a = ['foo', 'bar', 'baz', 'zzz']
while a:
    print(a.pop())  # a.pop(-2) 이러면 끝에서 2번째인 bar부터 나온다? 
    
    
# 3
n = 5
while n > 0:
    n -= 1
    if n == 2:
        continue
    print(n)
print('Loop Ended')

# 4
# While - else
m = 10
while m > 0:
    m -= 1
    print(m)
else:
    print('else out')
    
# 5
a = ['foo', 'bar', 'baz', 'qux']
s = 'zz'

i = 0
while i < len(a):
    if a[i] == s:
        break
    i += 1
else:
    print(s, 'not found in list')
    
    
# 6
b = ['foo', 'bar', 'baz', 'qux']
while True:
    if not b:
        break
    print(b.pop())
```
